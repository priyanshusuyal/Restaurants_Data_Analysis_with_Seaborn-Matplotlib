```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('indian_restaurants.csv')
```

```python
df
```

```python
df.rename(columns={'south_indian_or_not':'South_Indian'},inplace=True)
df.rename(columns={'north_indian_or_not':'North_Indian'},inplace=True)
df.rename(columns={'fast_food_or_not':'Fast_Food'},inplace=True)
df.rename(columns={'biryani_or_not':'Biryani'},inplace=True)
df.rename(columns={'bakery_or_not':'Bakery'},inplace=True)

df.rename(columns={'street_food':'Street_Food'},inplace=True)
```

```python
# Average Price Distribution

# Set your custom colors for the gradient and bar outline
color1 = '#7437E7'  # Purple
color2 = '#CA22A3'  # Pink
outline_color = '#A3FBF9'

# Create a custom colormap with two colors
custom_cmap = plt.cm.colors.LinearSegmentedColormap.from_list('custom', [color1, color2], N=256)

# plot
plt.figure(figsize=(10, 4))
Average_price_plot= sns.histplot(df['average_price'], bins=30, kde=True, color='#18E2F6', edgecolor=outline_color)

# Create a gradient using the custom colormap
gradient = np.linspace(0, 1, 35).reshape(1, -1).repeat(10, axis=0)

# Overlay the gradient on the histogram
Average_price_plot.imshow(gradient, aspect='auto', extent=Average_price_plot.get_xlim() + Average_price_plot.get_ylim(), origin='upper', alpha=0.8, cmap=custom_cmap)

# Set the title and labels
plt.title('Average Price Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Average Price',fontsize=10,fontweight='bold')
plt.ylabel('No. of Restaurants', fontsize=10 ,fontweight='bold')

plt.show()
```

```python

# Distribution of Cuisine Types

# Set your custom colors for the gradient and bar outline
color1 = '#7437E7'  # Purple
color2 = '#CA22A3'  # Pink
outline_color = '#A3FBF9'

#------------------------------------------------------------------------
# plot
df['cuisine_type']=df[['South_Indian','North_Indian','Fast_Food','Street_Food','Biryani','Bakery']].idxmax(axis=1)
# for eg.idxmax=[1,0,1,0,0,1]=>index(0,2,5)=>south_indian=1,north=0,fast=1.........
plt.figure(figsize=(12,5))
cuisine_type_plot=sns.countplot(x='cuisine_type', data=df,color='#18E2F6', edgecolor=outline_color,width=0.25)


for i in cuisine_type_plot.containers:
    cuisine_type_plot.bar_label(i)
    
plt.title('Distribution of Cuisine Types',fontsize=16,fontweight='bold')
plt.xlabel('Cuisine Type',fontsize=10,fontweight='bold')
plt.ylabel('No.of Restaurant',fontsize=10,fontweight='bold')

#------------------------------------------------------------------------
# Create a gradient using the custom colormap
gradient = np.linspace(0, 1, 35).reshape(1, -1).repeat(10, axis=0)
# Overlay the gradient on the histogram
cuisine_type_plot.imshow(gradient, aspect='auto', extent=cuisine_type_plot.get_xlim() + cuisine_type_plot.get_ylim(), origin='upper', alpha=0.8, cmap=custom_cmap)

# Set the title and labels
plt.title('Distribution of Cuisine Types', fontsize=16, fontweight='bold')
plt.xlabel('Cuisine Type')
plt.ylabel('No. of Restaurants')

plt.show()
```

```python
# Rating vs cusion type
# Set your custom colors for the gradient and bar outline
color1 = '#7437E7'  # Purple
color2 = '#CA22A3'  # Pink
outline_color = '#A3FBF9'

# plot
# --------------------------------------------------------------
a=['#18E2F6','#18E2F6','#18E2F6','#18E2F6','#18E2F6','#18E2F6']
plt.figure(figsize=(12,5))
rating_cusion =sns.boxplot(x='cuisine_type', y='rating', data=df, palette=a,hue='cuisine_type')
# ----------------------------------------------------------------
# Create a gradient using the custom colormap
gradient = np.linspace(0, 1, 35).reshape(1, -1).repeat(10, axis=0)
# Overlay the gradient on the histogram
rating_cusion.imshow(gradient, aspect='auto', extent=rating_cusion.get_xlim() + rating_cusion.get_ylim(), origin='upper', alpha=0.8, cmap=custom_cmap)

# Set the title and labels
plt.title('Rating Vs cuisine_type', fontsize=16, fontweight='bold')
plt.xlabel('Cuisine Type',fontsize=10,fontweight='bold')
plt.ylabel('Rating',fontsize=10,fontweight='bold' )
plt.show()
```

```python
# Location-wise Analysis:
# Explore the distribution of restaurants based on their locations.
# Set your custom colors for the gradient and bar outline
color1 = '#7437E7'  # Purple
color2 = '#CA22A3'  # Pink

# plot
#------------------------------------------------------------------------
plt.figure(figsize=(13,4))
order=df['location'].value_counts().head(12).reset_index()
order_plt=sns.pointplot(y='count',x='location',data=order,palette='viridis',hue='location')
for i in order_plt.containers:
 order_plt.bar_label(i)
#------------------------------------------------------------------------
# Create a gradient using the custom colormap
gradient = np.linspace(0, 1, 35).reshape(1, -1).repeat(10, axis=0)
# Overlay the gradient on the histogram
order_plt.imshow(gradient, aspect='auto', extent=order_plt.get_xlim() + order_plt.get_ylim(), origin='upper', alpha=0.8, cmap=custom_cmap)
# Set the title and labels

plt.title('Distribution of Top 10  Restaurants Based on There Locations', fontsize=16, fontweight='bold')
plt.xlabel('City',fontsize=10,fontweight='bold')
plt.ylabel('No. of Restaurants',fontsize=10,fontweight='bold' )

plt.show()
```

```python
# Set your custom colors for the gradient and bar outline
color1 = '#7437E7'  # Purple
color2 = '#CA22A3'  # Pink
outline_color = '#A3FBF9'

# Create a custom colormap with two colors
custom_cmap = plt.cm.colors.LinearSegmentedColormap.from_list('custom', [color1, color2], N=256)

# Plot
top_10_res = df[['restaurant_name', 'rating']].sort_values(by=('rating'), ascending=False).head(10)
plt.figure(figsize=(12,4))
top_10_res_plot = sns.barplot(x='restaurant_name', y='rating', data=top_10_res, color='#18E2F6', edgecolor=outline_color,width=0.4)
for i in top_10_res_plot.containers:
    top_10_res_plot.bar_label(i, fmt='%.2f', fontsize=10)

# Create a gradient using the custom colormap
gradient = np.linspace(0, 1, 35).reshape(1, -1).repeat(1, axis=1)

# Overlay the gradient on the histogram
top_10_res_plot.imshow(gradient, aspect='auto', extent=top_10_res_plot.get_xlim() + top_10_res_plot.get_ylim(), origin='upper', alpha=0.8, cmap=custom_cmap)

plt.xlabel('Restaurant Name')
plt.ylabel('Rating')
plt.title('Top 10 Restaurant by Rating', fontsize=15, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Assuming top_10_res is your DataFrame containing the data
# Ensure that the column names are consistent with your actual data

categories = list(top_10_res['restaurant_name'])
values = list(top_10_res['rating'])

# Create a colormap with a different color for each category
colors = plt.cm.viridis(np.linspace(0, 1, len(categories)))

plt.figure(figsize=(10, 6))
ax = plt.subplot(111, polar=True)

# Plot each category with a different color
for i in range(len(categories)):
    ax.fill_between([i * 2 * np.pi / len(categories), (i + 1) * 2 * np.pi / len(categories)],
                    0, values[i], alpha=0.8, color=colors[i], label=categories[i])

# Set radial tick labels
ax.set_yticklabels([])

# Set angular tick labels
ax.set_xticks(np.linspace(0, 2 * np.pi, len(categories), endpoint=False))
ax.set_xticklabels(categories)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
plt.title('Top 10 Restaurant by Rating', fontsize=15, fontweight='bold')
plt.legend(loc='upper right', bbox_to_anchor=(-0.2, 1))
plt.show()
```
